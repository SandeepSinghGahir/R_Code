```r
norm2 <- function(n,meanv,sdl,sdh){
    a <- rnorm(2*n,meanv,sdl)
    b <- rnorm(2*n*sdh/sdl,meanv,sdl)
    a <- a[a<meanv & a>0] 
    b <- b[b>=meanv]
    big <- c(a,b)
    sample(big,n)
}
```

```r
#Random Points
RP <- 1000

#Fixed Cost function
fixedl <- list(
list(10e6,500e3,2e6,"eng"),
list(3e6,100,100,"facilities"),
list(5e6,500e3,1e6,"mkt"),
list(6e6,1e6,2e6,"manf"),
list(1e6,100,500e3,"admin")
)

fixed_costs <- function(){
    wk <- rep(0,RP)
    for (l in fixedl){
        wk <- wk + norm2(RP,l[[1]],l[[2]],l[[3]])
    }
    wk
}

hist(fixed_costs(),100)
```

```r
FC <- fixed_costs()
print(summary(FC))
```

```r
FC <- fixed_costs(fixedl)
print(summary(FC))
```

```r
varl <- list(
list(20,2,5,"manf"),
list(0.50,0.10,0.20,"QA"),
list(1.5,0.1,2.0,"sales"),
list(0.30,0.1,0.1,"inventory")
)
```

```r
fixed_costs <- function(fixedl){
    wk <- rep(0,RP)
    for (l in fixedl){
        wk <- wk + norm2(RP,l[[1]],l[[2]],l[[3]])
    }
    wk
}
```

```r
FC <- fixed_costs(fixedl)
print(summary(FC))
```

```r
FC <- fixed_costs(varl)
print(summary(FC))
```

```r
FC <- fixed_costs(fixedl)
VC <- fixed_costs(varl)
ASP <- norm2(RP,40,15,5)
VOL <- norm2(RP,10e6,500e3,200e3)
REVENUE <- ASP*VOL
EXPENSES <- FC+VC*VOL
PROFIT <- REVENUE-EXPENSES
hist(PROFIT,100)
MakesMoney <- lapply(PROFIT,function(x){if (x > 0) 1 else 0})
print(sum(unlist(MakesMoney)))
```
