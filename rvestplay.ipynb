Read in the sjsu main website, and find all the links. The links are those things that are html 'a' use html_nodes to find those. The link portion is an attribute 'href' will use html_attr to find that.

```r
url<- "http://www.sjsu.edu"
sjsupage <- read_html(url)
nodes <- html_nodes(sjsupage,'a')
nodes
```

```r
html_attr(nodes,'href')
```

```r
html_attr(nodes,'href') -> links
nonsjsu <- grep('sjsu.edu',links,invert=TRUE)
nonsjsu
```

```r
links[nonsjsu]
```

This can be coded using the magrittr library pipe function %>%

A pipe replaces the first function argument with the results of the previous function call. If you want the data in another position, use the period '.' The above could have been coded as:

```r
read_html(url) %>% html_nodes('a') %>% 
html_attr(.,'href') -> links 
grep('sjsu.edu',links,invert=TRUE) %>% links[.]
```
