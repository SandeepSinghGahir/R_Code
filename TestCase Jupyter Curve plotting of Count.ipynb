```r
x <- c(1:100)
y <- x*x
x2 <- data.frame(x=x,y=y)
head(x2)
```

```r
mx2 <- lm(y ~ x,x2)
```

```r
plot(fitted(mx2,x2))
lines(x2,col="blue")
```

```r
mod2 <- lm(y ~ I(x^2),x2)
```

```r
plot(fitted(mod2,x2))
lines(x2,col="blue")
```

```r
mod3 <- lm(y ~ poly(x,2),x2)
plot(fitted(mod2,x2))
lines(x2,col="blue")
```

```r
a <- c(1:100)
b <- log(a)
lf <- data.frame(x=a,y=b)
mlog <- lm(y ~ x,lf)
plot(lf)
lines(lf$x,fitted(mlog),col="blue")
```

```r
a <- c(1:100)
b <- log(a)
#
q <- c(50:200)
r <- log(q)
tf <- data.frame(x=q,y=r)
lf <- data.frame(x=a,y=b)
mlog <- lm(y ~ I(log(x)),lf)
plot(tf)
lines(tf$x,predict(mlog,tf),col="blue")
```
