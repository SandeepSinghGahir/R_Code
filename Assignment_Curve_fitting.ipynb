```r
#Loading GM stock price data
GM <- read.csv("data/GM.csv")

#Converting GM$Date to Date datatype
GM$Date <- as.Date(as.character(GM$Date))

#Creating test data set out of GM data (all the objects before Dec 31 2015)
GMTrain <- GM[GM$Date < as.Date("2015-12-31"),]

#Creating common function for all the models
plotAll <- function(deg) {
    #Creating models for Low,Open and close prices of stock
    lowModel <- lm(Low ~ poly(Date,deg),GMTrain)
    openModel <- lm(Open ~ poly(Date,deg),GMTrain)
    closeModel <- lm(Close ~ poly(Date,deg),GMTrain)
    
    
    #plotting all the values of Low with date
    plot(GM$Date,GM$Low,col="blue",type="l")
    
    #Adding line curve for the predicted values
    lines(GM$Date,predict(lowModel,GM),col="Red",type="l")
    
    #Similarly creating plot and adding lines for predicted values of Open and Close stock value
    
    plot(GM$Date,GM$Open,col="blue",type="l")
    lines(GM$Date,predict(openModel,GM),col="Brown",type="l")
    
    plot(GM$Date,GM$Close,col="blue",type="l")
    lines(GM$Date,predict(closeModel,GM),col="Green",type="l")

}
```

```r
plotAll(3)
```
